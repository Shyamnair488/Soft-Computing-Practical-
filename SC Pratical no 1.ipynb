```python
A=dict()
B=dict()
Y=dict()

A={"a":0.3,"b":0.7,"c":0.4,"d":0.5,"e":0.8}
B={"a":0.6,"b":0.2,"c":0.9,"d":0.2}
print("Fuzzy Set A = ",A,"\nFuzzy Set B = ",B)
```

**For Union**

```python
#for union we use maximum
for A_key , B_key in zip(A,B):
  A_value=A[A_key]
  B_value=B[B_key]
  if A_value>B_value:
    Y[A_key]=A_value
  else:
    Y[B_key]=B_value

print("Fuzzy Set Of Union Is = ",Y)
```

**For Intersection**

```python
#for intersection we use minimum
for A_key , B_key in zip(A,B):
  A_value=A[A_key]
  B_value=B[B_key]
  if A_value<B_value:
    Y[A_key]=A_value
  else:
    Y[B_key]=B_value

print("Fuzzy Set Of Intersection Is = ",Y)
```

**For Complement**

```python
for A_key in A:
  Y[A_key]=1-A[A_key]
print("Fuzzy Set Of Complement Is = ",Y)
```

```python
for B_key in B:
  Y[B_key]=1-B[B_key]
print("Fuzzy Set Of Complement Is = ",Y)
```

**For Difference**

```python
#for difference we use min of (A,1-B)
for A_key , B_key in zip(A,B):
  A_value=A[A_key]
  B_value=1-B[B_key]
  if A_value<B_value:
    Y[A_key]=A_value
  else:
    Y[B_key]=B_value

print("Fuzzy Set Of Difference Is = ",Y)
```
