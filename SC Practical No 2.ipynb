```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

# Input universe
x = np.linspace(0, 10, 100)

# Triangular Membership Function
triangular_mf = fuzz.trimf(x, [3, 6, 9])

# Trapezoidal Membership Function
trapezoidal_mf = fuzz.trapmf(x, [2, 4, 7, 9])

# Gaussian Membership Function
gaussian_mf = fuzz.gaussmf(x, 5, 1.5)

# Bell Membership Function
bell_mf = fuzz.gbellmf(x, 2, 3, 7)

# Sigmoid Membership Function
sigmoid_mf = fuzz.sigmf(x, 5, 1)

# Generalized Bell Membership Function
generalized_bell_mf = fuzz.gbellmf(x, 2, 3, 8)

# Singleton Membership Function
singleton_mf = fuzz.trimf(x, [5, 5, 5])

# Pi-shaped Membership Function
pi_shaped_mf = np.fmax(fuzz.trapmf(x, [2, 4, 6, 8]), fuzz.trapmf(x, [4, 6, 8, 10]))

# Z-shaped Membership Function
z_shaped_mf = fuzz.trimf(x, [3, 6, 8])

# S-shaped Membership Function
s_shaped_mf = fuzz.trimf(x, [3, 5, 8])

# Plot membership functions
plt.figure(figsize=(12, 8))

plt.subplot(3, 3, 1)
plt.plot(x, triangular_mf, label='Triangular')
plt.legend()

plt.subplot(3, 3, 2)
plt.plot(x, trapezoidal_mf, label='Trapezoidal')
plt.legend()

plt.subplot(3, 3, 3)
plt.plot(x, gaussian_mf, label='Gaussian')
plt.legend()

plt.subplot(3, 3, 4)
plt.plot(x, bell_mf, label='Bell')
plt.legend()

plt.subplot(3, 3, 5)
plt.plot(x, sigmoid_mf, label='Sigmoid')
plt.legend()

plt.subplot(3, 3, 6)
plt.plot(x, generalized_bell_mf, label='Generalized Bell')
plt.legend()

plt.subplot(3, 3, 7)
plt.plot(x, singleton_mf, label='Singleton')
plt.legend()

plt.subplot(3, 3, 8)
plt.plot(x, pi_shaped_mf, label='Pi-shaped')
plt.legend()

plt.subplot(3, 3, 9)
plt.plot(x, z_shaped_mf, label='Z-shaped')
plt.plot(x, s_shaped_mf, label='S-shaped')
plt.legend()

plt.show()
```
