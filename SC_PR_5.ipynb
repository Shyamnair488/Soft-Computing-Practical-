```python
population = [
    {'chromosome': "A", 'fitness': 5},
    {'chromosome': "B", 'fitness': 2},
    {'chromosome': "C", 'fitness': 0.5},
    {'chromosome': "D", 'fitness': 1.5},
    {'chromosome': "E", 'fitness': 1.5},
]
```

CANONICAL SELECTION

```python
def canonical_selection(population, num_parents):
    # Sort the population based on fitness in descending order
    population.sort(key=lambda x: x['fitness'], reverse=True)

    # Select the top 'num_parents' individuals for reproduction
    selected_parents = population[:num_parents]

    return selected_parents
```

```python
# Number of parents to select
num_parents_to_select = 1

# Perform canonical selection
selected_parents = canonical_selection(population, num_parents_to_select)

# Display the selected parents
print("Selected Parents:")
for parent in selected_parents:
    print(f"Chromosome: {parent['chromosome']}, Fitness: {parent['fitness']}")
```
